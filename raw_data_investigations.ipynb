## investigation of feasibility of individual judge models

```python
#investigating counts per year per judge
counts = merged2.groupby(['tracid','comp_year'],as_index=False)['idnproceeding'].count()
counts = counts[counts.idnproceeding>49]#only include year if at least 50 obs per year
first_yr = counts.groupby('tracid')['comp_year'].first()#list of first year per judge

```

```python
#how many judges have at least 500 cases? assuming we would need at least this much data 
#for individual judge over time models

tracid_500 = merged2.groupby('tracid').idnproceeding.count()>=500 #bool indicating whether judge has at least 500 cases
tracid_500 = tracid_500.index.values[tracid_500]#indices of judges with at least 500 cases
merged3 = merged2.loc[merged2.tracid.isin(tracid_500)]

len(tracid_500)
```

```python
#of these, how many judges have cases that span at least 10 years?
judge_ranges = merged3.groupby('tracid')['comp_year'].apply(np.ptp)
print(len(judge_ranges[judge_ranges>10])) #323 out of 371 judges have at least 10 years of cases

```

## look at cases with multiple proceedings per case

```python

bla = merged2.groupby('idncase_x')['idnproceeding'].count()>1
cases = merged2.groupby('idncase_x')['idnproceeding'].count()[bla].index.values
merged2_multiples = merged2.loc[merged2['idncase_x'].isin(cases)]
pd.options.display.max_rows = 500
merged2_multiples.sort_values(['idncase_x','comp_date'])
```
